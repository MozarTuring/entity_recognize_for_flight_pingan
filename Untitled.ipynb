```python
with open('nohup.out', 'r') as rf:
    lines = rf.readlines()
import re
pat = re.compile(r'\d\.\d{1,4}')
ls = []
ls_t = []
for line in lines:
    if "Complete ocr in " in line:
        ls.extend(pat.findall(line))
    if "Time consumed: " in line:
        ls_t.extend(pat.findall(line))
        
new_ls = list(map(lambda x:float(x), ls))
new_ls_t = list(map(lambda x:float(x), ls_t))
print(len(new_ls), sum(new_ls)/len(new_ls))
print(len(new_ls_t), sum(new_ls_t)/len(new_ls_t))
```
