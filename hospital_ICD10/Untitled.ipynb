```python
def match_with_standard(inp_sent):
    data_sent = {"id":"123", "texts":[inp_sent], "is_tokenized":False}
    while True:
        try:
            response = requests.post("http://mjw-bert-as-service-gpu-stg.icore-aas-rbt-gpu-stg.paic.com.cn/encode", data=json.dumps(data_sent), headers={"Content-Type": "application/json"})
        except (requests.exceptions.SSLError, requests.exceptions.ConnectionError) as e:
            if "10054" in str(e) or "104" in str(e):
                continue
            else:
                raise Exception(e)
        break
    response_dic = json.loads(response.text)
    x = np.array(response_dic["result"][0])
    similarity_ls = []
    count = 0
    for k, v in dic.items():
    #     if count > 10:
    #         break
        count += 1
        y = np.array(v)
        similarity = np.dot(x,y)/(np.linalg.norm(x)*np.linalg.norm(y))
        similarity_ls.append((k,similarity))
    similarity_ls.sort(key=lambda x:x[-1], reverse=True)
    print(similarity_ls[:10])
```

```python
match_with_standard("左胫骨平台骨折(SCHATZKERII中医诊断:损伤骨折型)气滞血瘀证")
```
